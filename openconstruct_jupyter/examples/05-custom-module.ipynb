# 05 — Custom Module

Build a sense module from scratch.

```python
import openconstruct_jupyter as oc

client = oc.connect(notebook=True)
```

```python
def summarize_text(text: str) -> str:
    """A custom sense module that summarizes text."""
    words = text.split()
    return f'Summary: {len(words)} words, first 3: {" ".join(words[:3])}...'

oc.from_function(summarize_text, name='summarizer')
print('Registered summarizer tool!')
```
